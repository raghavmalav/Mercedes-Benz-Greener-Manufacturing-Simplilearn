```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
train_var = df_train.var()
train_var
```

```python
train_var_0 = train_var[train_var==0]
train_var_0
```

```python
train_var_0.shape
```

```python
train_var_0.index
```

```python
for index in train_var_0.index:
    if index in df_train.columns:
        df_train.drop([index],axis=1,inplace=True)
```

```python
for index in train_var_0.index:
    if index in df_test.columns:
        df_test.drop([index],axis=1,inplace=True)
```

```python
df_test.shape
```

```python
df_train.duplicated().value_counts()
```

```python
df_test.duplicated().value_counts()
```

```python
new_train_df = df_train.drop(['ID'],axis = 1)
```

```python
new_test_df = df_test.drop(['ID'],axis=1)
```

```python
new_train_df.isnull().sum().sum()
```

```python
new_test_df.isnull().sum().sum()
```

```python
target = new_train_df['y']
```

```python
target.head()
```

```python
new_train_df = new_train_df.drop(['y'],axis=1)
```

```python
new_train_df.head()
```

```python
DF = pd.concat([new_train_df,new_test_df])
```

```python
DF.shape
```

```python
cat_df = DF.select_dtypes(include=['object'])
```

```python
cat_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_df = cat_df.apply(LabelEncoder().fit_transform)
```

```python
cat_df.head()
```

```python
for col in cat_df.columns:
    if col in DF.columns:
        DF.drop([col],axis=1,inplace=True)
```

```python
DF.head()
```

```python
DF = pd.concat([DF,cat_df],axis=1)
```

```python
DF.head()
```

```python
corr_mat = DF.corr()
corr_mat
```

```python
upper_tri = corr_mat.where(np.triu(np.ones(corr_mat.shape),k=1).astype(np.bool))
```

```python
upper_tri.head()
```

```python
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
to_drop
```

```python
for col in DF.columns:
    if col in to_drop:
        DF = DF.drop([col],axis=1)
```

```python
DF.shape
```

```python
final_train_x = DF[:len(new_train_df)]
```

```python
final_test = DF[len(new_train_df):]
```

```python
final_train_x.shape
```

```python
final_test.shape
```

```python
target.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA()
x_train = pca.fit_transform(final_train_x)
x_test = pca.transform(final_test)
```

```python
import xgboost as xgb
```

```python
xgb_clf = xgb.XGBRegressor(
    n_estimators=100,
    reg_lambda=1,
    gamma=0,
    max_depth=3
)
```

```python
xgb_clf.fit(x_train,target)
```

```python
y_pred = xgb_clf.predict(x_test)
```
